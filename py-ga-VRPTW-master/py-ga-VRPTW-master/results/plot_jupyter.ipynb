```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os

# Check if the file exists
file_path = 'C204_uC8.0_iC100.0_wC1.0_dC1.5_iS100_pS400_cP0.85_mP0.02_nG300.csv'
if os.path.exists(file_path):
    # Read CSV file
    df = pd.read_csv(file_path)

    # Specify columns to plot
    columns_to_plot = ['min_fitness','max_fitness','avg_fitness','std_fitness'
]  # Add more columns as needed
else:
    print(f"File {file_path} does not exist. Please check the file path and name.")

```

```python
plt.figure(figsize=[8, 8])
df[columns_to_plot].plot()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Columns Plot')
plt.legend(loc='lower right')
plt.grid(True)
plt.xlim(0)
plt.ylim(0)

plt.savefig(rf'{file_path}'+'.jpeg')
```

```python
# Plotting specified columns
for column in range(len(df.columns)):
    plt.plot(df[column], label=column)

# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Selected Columns Plot')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
